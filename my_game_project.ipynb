# Identification of patterns that determine the success of the game


## Introduction

### Project description

The online store "Streamchik" sells computer games all over the world. Historical data on game sales, user and expert ratings, genres and platforms (for example, Xbox or PlayStation) are available from open sources. It is necessary to identify patterns that determine the success of computer games. This will allow you to bet on a potentially popular product and plan advertising campaigns.

There is access to data until 2016. Let's imagine that it's December 2016, and an advertising campaign is being planned for 2017. We need to work out the principle of working with data. It does not matter whether sales are forecast for 2017 according to 2016 data or for 2027, based on 2026 data.

The data set contains the abbreviation ESRB (Entertainment Software Rating Board), an association that determines the age rating of computer games. ESRB evaluates game content and assigns it a suitable age category, for example, "For adults", "For younger children" or "For teenagers".

**MAIN GOAL - to find out key parameters influencing the sales the most for modern games**

### Data description

- `Name` — name of the game
- `Platform` — platform 
- `Year_of_Release` — year of release
- `Genre` — game genre
- `NA_sales` — sales in North America (millions of copies sold)
- `EU_sales` — sales in Europe (millions of copies sold)
- `JP_sales` — sales in Japan (millions of copies sold)
- `Other_sales` — sales in other countries (millions of copies sold) 
- `Critic_Score` — critics' rating (maximum 100)
- `User_Score` — user rating (maximum 10)
- `Rating` — rating from the ESRB organization (Eng. Entertainment Software Rating Board). This association determines the rating of computer games and assigns them a suitable age category.

ESRB rating definitions:
- `E`: for everyone (for all ages)
- `M`: for matures and older (15+ y.o.)
- `T`: for teens and older (13+ y.o.)
- `E10+`: for 10+ y.o.
- `K-A`: Kids to Adults. Was used until 1998. Changed to E rating
- `AO`: Adults Only (18+ y.o.)
- `EC`: Early Childhood. Games with this rating contain content which is aimed towards a preschool audience.This rating was deprecated in 2018 due to rarity of use.
- `RP`: Rating Pending. Awaiting for final rating.

```python
# identifying outliers with IQR

q75,q25 = np.percentile(data.total_sales,[75,25])
intr_qr = q75-q25
max_val = q75+(1.5*intr_qr)
min_val = q25-(1.5*intr_qr)
```

```python
# big outliers

display(data[data['total_sales'] > max_val].head(10))
print('Number of rows:', data[data['total_sales'] > max_val].count()[0])
```

```python
# small outliers

display(data[data['total_sales'] < min_val].head(10))
print('Number of rows:', data[data['total_sales'] < min_val].count()[0])
```

```python
# calculating number of rows and proportions for outliers

def num_prop_outliers(data):
    print('Number of outliers:',data[data['total_sales'] > max_val].shape[0], 
      ', proportion of outliers inside of the period:',
      round(data[data['total_sales'] > max_val].shape[0]/data.shape[0]* 100,2), '%' 
     )
```

```python
# for the recent period

num_prop_outliers(data_recent)
```

```python
outliers = data[data['total_sales'] > max_val]
outliers['year_of_release'].value_counts().sort_index()
```

**Description:**

It can be seen that there are now tiny outliers that a smaller than min_value calculated with IQR. But there are some games that are bigger than the max_value threshold and the number of games in 1869. That is massive amount of data that contains valuable info about successful games like Super Mario Bros.

The most outliers is concentrated in the second period. It may be connected with the fact that the biggest number of games was released in the same period.

**Descision:**

Let's have 2 datasets: with outliers (game hits with big total sales) and filtered data without outliers, and compare the results with each other to estimate the effect of the outliers on the correlation.

```python
def outliers_filter(data):
    filtered_data = data[data['total_sales'] < max_val]
    return filtered_data
```

```python
data_relevant_agg_filtered = outliers_filter(data_relevant_agg)
data_relevant_agg_filtered.head()
```

```python
data_recent_filtered = outliers_filter(data_recent)
data_recent.head()
```

```python
# automation of correlation calculations and vizualisations

def corr_scatter(data,param1,param2):
    
    plt.figure(figsize=(16,8))
    sns.scatterplot(data=data, 
            x=param2, 
            y=param1,
            alpha=0.5)
    plt.title('Correlation of {} and {}'.format(param1,param2), fontsize=16)
    plt.xlabel('{}'.format(param2), fontsize=14)
    plt.ylabel('{}'.format(param1), fontsize=14)
    
    x_ticks = np.arange(0,11,1)
    plt.xticks(x_ticks)
    
    plt.show()
    
    
    print('Correlation:', round(data[param1].corr(data[param2]),3))
```
